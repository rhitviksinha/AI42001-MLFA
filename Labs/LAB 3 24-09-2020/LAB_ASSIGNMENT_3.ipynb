### Part 1

*Construct the vocabulary without stop-words.*

```python
train_sentences = []
train_labels = []

test_sentences = []
test_labels = []

for row in traindata:
    train_labels.append(row[0])
    train_sentences.append(row[1])

for row in testdata:
    test_labels.append(row[0])
    test_sentences.append(row[1])

print("Train Sentences:", train_sentences[:5])
print("Train Labels:", train_labels[:5])
print()
print("Test Sentences:", test_sentences[:5])
print("Test Labels:", test_labels[:5])
```

```python
# remove stop words from sentence

import nltk
import string
# nltk.download('stopwords')
# nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def process_text(sentence):
    STEM = PorterStemmer()
    STOPWORDS_EN = stopwords.words("english")
    tokens = word_tokenize(sentence)
    processed_text = []
    for word in tokens:
        if (word not in STOPWORDS_EN and word not in string.punctuation):  # remove stopwords and punctuation
            word = word.lower()
            word_stem = STEM.stem(word)  # stem word
            processed_text.append(word_stem)
    return processed_text

# test process_text() with a sample sentence

trial = "Hi! My name is Rhitvik Sinha. My roll number is '18EE10043'. Nice to meet you!"
processed_text = process_text(trial)
print(processed_text)
```

```python
# construct vocabulary from a list of (list of words) provided for training

processed_train_data = []
for sentence in train_sentences:
    processed_train_data += process_text(sentence)

from sklearn.feature_extraction.text import CountVectorizer
corpus = CountVectorizer().fit(processed_train_data)

corpus_freqs = corpus.vocabulary_
print(corpus_freqs)
print(len(corpus_freqs))
```

### Part 2

*Calculate the prior distribution of the labels.*

```python
prior_science, prior_sports, prior_business, prior_covid = 0.0, 0.0, 0.0, 0.0
N = float(len(train_labels))

for label in train_labels:
    if label == "science":
        prior_science += 1.0 / N
    elif label == "business":
        prior_business += 1.0 / N
    elif label == "sports":
        prior_sports += 1.0 / N
    elif label == "covid":
        prior_covid += 1.0 / N
    else:
        print("ERROR: " + label)

prior_business = round(prior_business, 4)
prior_covid = round(prior_covid, 4)
prior_science = round(prior_science, 4)
prior_sports = round(prior_sports, 4)

print("prior(business) = {}".format(prior_business))
print("prior(covid)    = {}".format(prior_covid))
print("prior(science)  = {}".format(prior_science))
print("prior(sports)   = {}".format(prior_sports))
```

### Part 3

*Calculate the class-conditional probabilities of each word in the vocabulary, for each topic.*

```python
processed_text_business = []
processed_text_covid = []
processed_text_science = []
processed_text_sports = []

for index, sentence in enumerate(train_sentences):
    if train_labels[index] == "business":
        processed_text_business += process_text(sentence)
    elif train_labels[index] == "covid":
        processed_text_covid += process_text(sentence)
    elif train_labels[index] == "science":
        processed_text_science += process_text(sentence)
    elif train_labels[index] == "sports":
        processed_text_sports += process_text(sentence)
    else:
        print("ERROR: " + train_labels[index])

business_corpus = CountVectorizer().fit(processed_text_business)
covid_corpus = CountVectorizer().fit(processed_text_covid)
science_corpus = CountVectorizer().fit(processed_text_science)
sports_corpus = CountVectorizer().fit(processed_text_sports)

business_corpus_freqs = business_corpus.vocabulary_
covid_corpus_freqs = covid_corpus.vocabulary_
science_corpus_freqs = science_corpus.vocabulary_
sports_corpus_freqs = sports_corpus.vocabulary_

for word, corpus_freq in corpus_freqs.items():
    if word not in business_corpus_freqs:
        business_corpus_freqs[word] = 0
    if word not in covid_corpus_freqs:
        covid_corpus_freqs[word] = 0
    if word not in science_corpus_freqs:
        science_corpus_freqs[word] = 0
    if word not in sports_corpus_freqs:
        sports_corpus_freqs[word] = 0

print("Business:")
print(business_corpus_freqs)
print(str(len(business_corpus_freqs)) + "\n")

print("Covid:")
print(covid_corpus_freqs)
print(str(len(covid_corpus_freqs)) + "\n")

print("Science:")
print(science_corpus_freqs)
print(str(len(science_corpus_freqs)) + "\n")

print("Sports:")
print(sports_corpus_freqs)
print(len(sports_corpus_freqs))
```

```python
n_business, n_covid, n_science, n_sports = 0, 0, 0, 0

for i in business_corpus_freqs.values():
    n_business += i

for i in covid_corpus_freqs.values():
    n_covid += i

for i in science_corpus_freqs.values():
    n_science += i

for i in sports_corpus_freqs.values():
    n_sports += i
```

```python
business_cond_prob = {}
covid_cond_prob = {}
science_cond_prob = {}
sports_cond_prob = {}

for word, corpus_freq in corpus_freqs.items():
    business_cond_prob[word] = (1 + business_corpus_freqs[word]) / (n_business + len(corpus_freqs))     # Laplacian Smoothing
    covid_cond_prob[word] = (1 + covid_corpus_freqs[word]) / (n_covid + len(corpus_freqs))              # Laplacian Smoothing
    science_cond_prob[word] = (1 + science_corpus_freqs[word]) / (n_science + len(corpus_freqs))        # Laplacian Smoothing
    sports_cond_prob[word] = (1 + sports_corpus_freqs[word]) / (n_sports + len(corpus_freqs))           # Laplacian Smoothing
```

```python
print("Business Conditional Probabilities:")
print(business_cond_prob)
print(str(len(business_cond_prob)) + "\n")

print("Covid Conditional Probabilities:")
print(covid_cond_prob)
print(str(len(covid_cond_prob)) + "\n")

print("Science Conditional Probabilities:")
print(science_cond_prob)
print(str(len(science_cond_prob)) + "\n")

print("Sports Conditional Probabilities:")
print(sports_cond_prob)
print(len(sports_cond_prob))
```

### Part 4

*For each test sentence, create the posterior distribution over the labels.*

```python
print("FORMAT:\n------\nGiven Label - Prediction - Probability - Text\n---------------------------------------------")
ACCURACY = 0.0
test_len = len(test_labels)

for sentence, label in zip(test_sentences, test_labels):
    print()
    words = process_text(sentence)
    prob_business = []
    prob_covid    = []
    prob_science  = []
    prob_sports   = []
    for word in words:
        if word in corpus_freqs:
            prob_business.append(( (1 + business_corpus_freqs[word]) / (n_business + len(corpus_freqs)) ))
            prob_covid.append(( (1 + covid_corpus_freqs[word]) / (n_covid + len(corpus_freqs)) ))
            prob_science.append(( (1 + science_corpus_freqs[word]) / (n_science + len(corpus_freqs)) ))
            prob_sports.append(( (1 + sports_corpus_freqs[word]) / (n_sports + len(corpus_freqs)) ))

    likelihood_business = 1.0
    for element in prob_business:
        likelihood_business *= element
    posterior_business = prior_business * likelihood_business

    likelihood_covid = 1.0
    for element in prob_covid:
        likelihood_covid *= element
    posterior_covid = prior_covid * likelihood_covid
    
    likelihood_science = 1.0
    for element in prob_science:
        likelihood_science *= element
    posterior_science = prior_science * likelihood_science
    
    likelihood_sports = 1.0
    for element in prob_sports:
        likelihood_sports *= element
    posterior_sports = prior_sports * likelihood_sports

    posterior_total = posterior_business + posterior_covid + posterior_science + posterior_sports

    posterior_business /= posterior_total
    posterior_covid /= posterior_total
    posterior_science /= posterior_total
    posterior_sports /= posterior_total

    posterior = [round(posterior_business, 4), round(posterior_covid, 4), round(posterior_science, 4), round(posterior_sports, 4)]
    topic = ["business", "covid", "science", "sports"]
    probability = max(posterior)
    predicted_label = topic[posterior.index(probability)]

    print(label + " - " + predicted_label + " - " + str(probability) + " - " + sentence)

    if label == predicted_label:
        ACCURACY += 1.0 / test_len
```
